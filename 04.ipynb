```python
import pandas as pd
df = pd.read_csv('data/superstore.csv', encoding='ISO-8859-1')
df.groupby(['Customer ID', 'Customer Name'], as_index=False)['Sales'].sum().sort_values('Sales', ascending=False).head(10).rename(
    columns={
        "Customer ID": "ID Cliente",
        "Sales": "Total Compras",
        "Customer Name": "Nome Cliente"
    })
```
