```python
import pandas as pd
df = pd.read_csv('data/superstore.csv', encoding='ISO-8859-1')
df.groupby(['Customer ID', 'Customer Name'], as_index=False)['Order ID'].count().sort_values(
    'Order ID', ascending=False).head(10).rename(columns={
        'Customer ID': 'ID Cliente',
        'Customer Name': 'Nome Cliente',
        'Order ID': 'Qtd. Pedidos'
    })
```
