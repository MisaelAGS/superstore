```python
import pandas as pd
df = pd.read_csv('data/superstore.csv', encoding='ISO-8859-1')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Mes Pedido'] = df['Order Date'].dt.month
df.groupby('Mes Pedido', as_index=False)['Order ID'].count().rename(columns={
    'Mes Pedido': 'Mês Pedido',
    'Order ID': 'Qtd. Pedidos'})
```
